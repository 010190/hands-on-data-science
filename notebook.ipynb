##  Useful shortcuts to know
1. `control + a`-> moves cursor to beginning of the line
2. `control + e` -> moves cursor to the end of the line
3. `control + f` - > (f like in forward) moves cursor one char ahead
4. `control + b` -> (b like in back) moves cursor one character backwards



```python
def square(x):
    """square a number"""
    return x ** 2

for N in range(1, 4):
    print(f"{N} squared is {square(N)}")
```

## IPython Magic Commands
1. `run <filename>` executes file mentioned in command
2. ` %timeit <function>` returns time needed for execution of mentioned function <br> If you want to measure time of execution of multiline code, just add second % sign, as in `%%timeit`
3. `%magic` returns list of available magic functions


## Input and Output history
IPython enables accessing inputted and outputted things <br> By using `In` we can access list containing every line of code of the file, where every list item is dedicated to each line <br> `Out` works identically, it outputs list of outputs for each line <br>
We can access outputed value of specific cell by using `Out[<cell_num>]`, however it is important to insert correct cell number, otherwise we will receive `KeyError Exception`
To print last outputted value, we can simply use `print(_)`



```python
import math
math.sin(2)
```
