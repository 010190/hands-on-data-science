## Errors and Debugging
### Errors
IPython enables 3 modes of controlling exceptions. We can specify it by calling `xmode <mode>`, where mode can be:
1. `xmode Context` default exception handling procedure
2. `xmode Plain` shorter version, we do not receive as much info as with `context` mode on
3. `xmode Verbose` the most advanced mode, where we achieve all of the `context` mode information + info about arguments passed to the functions that were called. The downside of this mode is that, the text message can get enormously long when code is more complex

```python
%xmode Verbose
```

## Two's complement numbers
Two's complement number is a form of presenting integers in binary system. Logic of binary system is as follows - we present number as if it was made as a sum of different powers of number 2. Starting from the edge we have 2^0, which is 1. Then every following number have increased by 1 power: *2^1, 2^2,2^3 ...*. In such representation number 5 is presented as "101", because we have `1*2^2+0*2^1+1*2^0` <br>
When we want to represent negative numbers as well, we need to add small change,
 being that the first digit states if we have negative or positive number  (0 for
  positive, 1 for negative). Writing negative number -x is equal with writing (x-1) bitwise, with all the bits complemented (swapped from 0 to 1 and from 1 to 0). This way writing -10 bitwise is equal to writing complement(9)=complement("00001001")=11110110"


## Bitwise operators
Python contains several bitwise operators:
1. `x << y` returns x with the bits shifted to the left by y place (new bits on the right-hand are zeros). It is equal to multiplying x by 2**y
5 << 2
"101" = "10100" = 20 = 5 * (2^2)
2. `x >> y` returns x with bits shifted to the right by y place. This is the same as floor division of x by 2**y
3. `x&y` does a bitwise *and*. Each bit of the output is 1 if the corresponding bit of x AND y is 1, otherwise it is 0.
4.  `x|y` each bit of the output is 0 if the corresponding bit of x AND y is 0, otherwise ut us 1.
5. `~x` returns the complement of x - the number you get by switching 1 for 0 and each 0 for a 1. Tis is the same as -x-1
6. `x^y` does a bitwise exclusive or. Each bit of the output is the same as the corresponding bit in x if that bit in y is 0, and it's he complement of the bit in x if that bit in y is 1.
Infinite series of 1 bits is a negative number.

```python
print(~20)
```

## Line-by-line profiling with `%lprun`
Sometimes more useful is to profile each line of our function and get insights of it. We can do it with use of `%lprun`, however first we need to install the package `line_profiler` <br> `pip install line_profiler` and then we need to load it `%load_ext line_profiler` and we are ready to profile any of our functions line-by-line. <br> *Example* <br> `%lprun -f sum_of_lists(5000)` <br> the result of the function is report for every line of function. Time is measured in microseconds. Knowlage of detailed time consumption of each line is crucial, as it enables us to achieve better performance by changing some of the most time consuming code.

## Profiling memory use with `%memit` and `%mprun`
Anothe aspect of profiling is seeing how much memory an operation use. We can measure it with `memory_profiler` packagce. First we have to install it and load it. <br> `pip install memory_profiler`. <br> Next we load it `%load_ext memory_profiler`. Now we can get useful insight of our function memory consumption by simply calling `%memit sum_of_lists(100000)` <br> However to use `%mprun` we need to first create simple module containg our function, otherwise it wont work. `%% file mprun_demo.py`, then all we need to do is import our function from module `from mprun_demo import sum_of_lists` and call the `%mprun` --> `%mprun -f sum_of_lists sum_of_lists(1000000)`

## NumPy array
NumPy array is object similar to the Python list, however there are some differences between them. Python list can contain heterogeneous datatypes, but NumPy array can only contain same data type object. Python list is more flexible however also need much more resources, as every object have to contain additional informations about it, such as its type. NumPy array however, similar to C array can contain only homogenous objects.

```python
np.array([1,2,3,4], dtype=np.float32)
```

```python
np.array([1,2,3])
```

NumPy arrays also are multidimensional --> array can contain another array or arrays, while Python list is strictly one dimensional

```python
np.array([[range(i,i+3) for i in range(1,4)]])
```
